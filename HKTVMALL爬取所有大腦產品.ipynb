```python
# import related module
import requests
import pandas as pd
import time
import random
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 正式爬取區域


# check how many pages in the category 0-52 for i in range(0, 53)
# fetch the API payload parameters

file_name = '2024-05-07_hktvmall_brainHealthProduct'
url = "https://cate-search.hktvmall.com/query/products"

# 將 'products' 中的所有元素名稱作為欄位名稱
column = ['baseProduct','name', 'averageRating', 'brandName', 'countryOfOrigin','description','salesVolume','storeName','summary','packingSpec']  # 將你要的元素名稱加到這個list中



df_origin= pd.DataFrame(columns=column)

for i in range(0, 36): #should be 36
  payload = {
      'query': ':relevance:street:main:category:AA13306100001:',
      'currentPage': i,
      'pageSize': 60,
      'pageType': 'searchResult',
      'categoryCode': 'AA13306100001',
      'lang': 'zh',
      'CSRFToken': '66dd1301-9a7b-497d-a666-585318dac027'
  }
  try:
    response = requests.post(url, data=payload)
    if response.status_code == 200:
      print(f'Page {i+1} is respond')

      data = response.json()

      # 使用列表推導式和 zip() 函數來創建 DataFrame
      try:

        #df = pd.DataFrame({col: [item[col] for item in data['products']] for col in column})

        # 構建 DataFrame
        df_data = []

        for item in data['products']:
            row = {}
            for col in column:
                # 如果列存在，則取得該值，否則設置為默認值None
                row[col] = item.get(col, None)

            # 取得價格資訊並確認可與其他資料對應
            price_list = item.get('priceList', [])
            buy_price = None
            discount_price = None
            for price_item in price_list:
                if price_item.get('priceType') == 'BUY':
                    buy_price = price_item.get('value')
                elif price_item.get('priceType') == 'DISCOUNT':
                    discount_price = price_item.get('value')

            # 將價格資訊添加到行中
            row['BuyPrice'] = buy_price
            row['DiscountPrice'] = discount_price

            # 將該行資料添加到DataFrame中
            df_data.append(row)

        # 使用填充缺失值的方式來建立 DataFrame
        df = pd.DataFrame(df_data)
      except Exception as e:
        print('創建df出意外了', e)

      # 顯示 DataFrame
      df_origin = pd.concat([df_origin, df], axis = 0)
      print(f'第{i+1}輪，資料框大小為{df_origin.shape}')
      #
      df_origin.to_csv(f'/content/drive/My Drive/Colab Notebooks/{file_name}.csv')

      sleep_time = random.randint(15, 30)
      time.sleep(sleep_time)

  except Exception as e:
    print("發生異常:", e)

df_origin
```
